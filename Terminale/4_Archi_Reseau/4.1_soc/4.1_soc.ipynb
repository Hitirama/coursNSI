# 4.1 SoC "System on Chip"

## Histoire

### Première génération

La première génération d’ordinateurs date de la seconde guerre mondiale:
- Le [Z3](https://fr.wikipedia.org/wiki/Zuse_3) était un calculateur à relais électromagnétique conçu en Allemagne. Opérationnel en 1941, c'était la première machine programmable et peut donc être considéré comme le premier ordinateur. Composé de 2000 relais électromagnétiques il fonctionnait à une fréquence d’horloge d’environ 10Hz.
- [Colossus](https://fr.wikipedia.org/wiki/Colossus_(ordinateur)) est une série de calculateurs électroniques britanniques fondés sur le système binaire. Le premier le Colossus Mark I est opérationnel en 1943. Constitué de tubes à vide (ancêtre du transistor). il accomplissait 5000 opérations par seconde. Son application était le déchiffrement du code de Lorenz utilisé par les hauts dirigeants allemands pour crypter leurs communications.
- L’[ENIAC](https://fr.wikipedia.org/wiki/ENIAC#:~:text=L%27ENIAC%20(acronyme%20de%20l,principe%2C%20tous%20les%20probl%C3%A8mes%20calculatoires.) est le premier ordinateur entièrement électronique. Opérationnel en 1945, son utilisation était les calculs de balistique pour l’armée américaine. Il a servi jusqu’en 1955. Capacité 100000 additions par seconde, 357 multiplications/s ou 38 divisions/s

### Seconde génération : arrivée du transistor

<img src="img/4.1_transistor_bell.webp" width="25%" style="float:right" alt="une photographie du premier transistor" title="une photographie du premier transistor">

L’invention du transistor au laboratoire Bell en 1947 permet de remplacer les tubes à vide. Ses avantages sont:
- taille réduite
- fiabilité augmentée
- coût réduit


L' [IBM 1401](https://fr.wikipedia.org/wiki/IBM_1401) a été fabriqué de 1959 à 1965. Il faut le plus vendu des ordinateurs de seconde génération. Environ 10000 unités produites aux Etats unis.

[Vidéo sur l’IBM 1401](https://www.youtube.com/watch?v=xUQnVt5qS-Y)

### Troisième génération : les circuits intégrés

Les liaisons électriques multiples entre les transistors sont complexes et coûteuses à réaliser et sont souvent l’origine d’un manque de fiabilité.

Le premier développement fut de monter les transistors sur des plaques puis de graver les fils qui les relient. On appelle cela un [circuit imprimé](https://fr.wikipedia.org/wiki/Circuit_imprim%C3%A9).

La création de circuits intégrés, qui comportent dans un même ensemble les transistors, diodes, résistances et liaisons, date de 1959. Cependant le coût était alors trop élevé pour être commercialisé.
<img src="img/4.1_apollo.jpeg" width="25%" style="float:right" alt="Une photographie de l'Apollo guidance computer" title="Une photographie de l'Apollo guidance computer">


La course à la Lune entre Américains et Russes a nécessité la création d’ordinateurs assez petits pour pouvoir être embarqués.

L’[Apollo guidance computer](https://fr.wikipedia.org/wiki/Apollo_Guidance_Computer) est ainsi le premier ordinateur à utiliser des circuits intégrés.

Caractéristiques:

- multitâche (6 tâches en parallèles)
- architecture 16bits
- 72ko de mémoire morte (contenant tous les programmes)
- 4ko de mémoire vive
- processeur constitué de 4100 portes NOR, chacune étant un circuit imprimé composé de 8 transistors.
- 32kg


Les sources des programmes utilisés pour Apollo 11 sont [disponibles sur github](https://github.com/chrislgarry/Apollo-11/).


Les recherches effectuées ont permis de diminuer les coûts de fabrication des circuits intégrés. Ils se sont retrouvés présents dans les mini-ordinateurs mais également dans les appareils domestiques (téléviseurs, radios…).

Les ordinateurs les plus célèbres de cette troisième génération sont ceux de la famille IBM/360.

[Une vidéo de 1969 sur la miniaturisation des composants électroniques d’un ordinateur](https://enseignants.lumni.fr/export/player/00000000640)

### Quatrième génération : les microprocesseurs

Un [microprocesseur](https://fr.wikipedia.org/wiki/Microprocesseur) est un "boitier" dans lequel tous les composants ont été suffisament miniaturisés afin d'effecteur des instructions pour effectuer un tâche.

- 1968 : Fondation de Intel
- 1969 : Fondation d’AMD
- 1970 : Intel met au point la première puce à mémoire dynamique DRAM (1ko)
- 1971 : Intel met au point la première puce à mémoire programmable EPROM
- 1971 : Intel invente le premier microprocesseur : [Intel 4004](https://fr.wikipedia.org/wiki/Intel_4004)

  - 108KHz
  - 4 bits
  - 2300 transistors

<img src="img/4.1_Intel_4004.jpg" width="50%" alt="Intel 4004" title="Intel 4004">

- 1977 : Sortie de l’Apple II

<img src="img/4.1_apple_2.jpg" width="25%" alt="Apple II" title="Apple II">


  - 4ko de mémoire vive
  - 16ko de mémoire morte
  - 1 MHz
  - lecteur de disquette
  - 1295$
- 1978: Premier microprocesseur 16 bits le 8086
  - 29000 transistors de 3 microns
  - mémoire adressable de 1Mo
- 1981 : IBM sort l’ IBM PC (Personal Computer) disposant d’une architecture standardisée.

<div class="alert alert-info">
    Depuis cette époque les microprocesseurs ont gagné en puissance mais l’architecture interne reste la même : un modèle Von Neumann basé sur des microprocesseurs de plus en plus puissants.

On parle parfois d'ordinateurs de 5ème génération mais les progrès effectués sont surtout logiciels et non matériels.
    
</div>

<div class="alert alert-warning">
    Les constats de ces évolutions sont:
    
- une miniturisation linéaire
- une fiabilité accrue
- un coût réduit
- une intégration continue des différents composants
- une puissance de calcul croissante
- une diminution de la puissance (Watt) nécessaire

## le PC moderne

Dans un PC les composants sont réunis autour d'une carte mère:

On peut citer:

- le CPU (*Central Processing Unit*)  qui exécute les instructions machines des programmes informatiques;
- la mémoire vive ou RAM (*Random Acess Memory*) dans laquelle sont stockées temporairement les informations traitées par le processeur;
- le disque dur qui contient le système d'exploitation, les programmes et les données des utilisateurs.

<img src="img/4.1_motherboard.jpg">


Ce type d’ordinateur est basé sur la **performance de calcul**. Les cartes graphiques ainsi que les processeurs sont équipés de système de refroidissement (ventilation voire refroidissement liquide).
Les deux grands fabricants de processeurs équipant ce type d’ordinateurs sont *Intel* et *AMD*.

L’ère de la miniaturisation à également provoquée celle de l’embarqué. D’abord avec les ordinateurs portables (laptop), puis les smartphones et les nanos-ordinateurs comme le raspberry pi.

<img src="img/4.1_raspberry.jpg" alt="Raspberry pi 4" title="Raspberry pi 4" width="25%">

Une nouvelle problématique s’est alors posée : celle de l’**autonomie** des appareils.


## Les systèmes sur puce (Soc : System on Chip)

Depuis l'avènement de l’ère des smartphones et tablettes, la problématique de l’autonomie de ces appareils est cruciale. Le système de batteries a fait d’énormes progrès mais s’est également accompagné d’une **miniaturisation** et d’une **intégration** toujours plus grande des composants de ces appareils.

Les nouveaux procédés de gravure ont permis une réduction significative de la taille des composants.

### Différence entre processeur et SoC

Le SoC intègre un ou plusieurs processeurs mais également d’autres composants nécessaires au fonctionnement d’un smartphone, tablette ou montre connectée et tout ceci au sein d’une seule puce.

### Composants d’un SoC

<img src="img/4.1_soc.jpg" alt="SoC" title="SoC" width="50%">

#### Le processeur (CPU : Central Processing Unit)

Il joue le même rôle que celui d’un PC mais la différence fondamentale est qu’il n’est pas optimisé pour la performance de calcul, il est optimisé pour la **meilleure efficacité en termes d’énergie**. Ce principe est le même dans les ordinateurs portables.
- La fréquence est optimisée
- Le type de coeur
  - certains sont optimisés pour consommer le moins possible avec des performances réduites, d’autres ont des performances plus élevées pour les calculs intensifs.
  - On trouvera donc des coeurs différents au sein du même CPU
- La finesse de gravure est un facteur important pour la baisse de la consommation.

#### La puce graphique (GPU : Graphical Processing Unit)
Elle permet le calcul des pixels sur l’écran. Sur un écran Ultra HD (3840x2160) pour une fréquence de rafraîchissement de 60Hz, il faut donc calculer 3840x2160x60 pixels soit environ 500 millions de pixels par seconde !

#### Le processeur de signal numérique (DSP : Digital Signal Processor)
En charge du traitement des signaux numériques (compression décompression de musique, de vidéos)

#### Le processeur de signal d’images (ISP : Image Signal Processor)
En charge du traitement de la création d’images numériques provenant du capteur de la caméra. Il assure les post traitements nécessaires des photos pour contrebalancer la mauvaise qualité de l’optique.
#### Le processeur de sécurité (SPU : Security Processor Unit)
Alimentation électrique indépendante. Stockage des données biométriques, bancaires…
#### Le processeur de calcul neuronal (NPU : Neuronal Processor Unit)
Dédié à l’intelligence artificielle : reconnaissance vocale ou d’image.
#### La mémoire
La mémoire est également intégrée dans le SoC. On y trouve différents types ( ROM, RAM, EEPROM)
#### Interfaces
Les SoCs ont également des interfaces pour pouvoir communiquer avec:
- Wifi
- Bluetooth
- 3G/4G/5G ( parfois intégré)
- radio FM
- ...

La communication entre les différents composants du SoC est assurée par un système traditionnel de Bus.

Des recherches sont actuellement en cours sur le **NoC** (Network on chip) qui permettrait de créer un réseau IP entre les différents composantes afin de limiter le goulot d’étranglement que constitue le BUS. ( [Voir article très pointu](https://www.design-reuse.com/articles/10496/a-comparison-of-network-on-chip-and-busses.html))
On parle également de ONoC (Optical Network on Chip). Voir le processeur [Hummingbird](https://www.lightelligence.ai/index.php/product/hummingbird.html).

### L'architecture ARM des SoC

Les PCs sont basés sur l’architecture x86 d’Intel, la quasi totalité des smartphones sont, quant à eux, basés sur l’architecture **ARM**.
De conception plus simple et moins énergivores, ils sont devenus dominants dans le secteur de l’embarqué, domaine délaissé en son temps par Intel et AMD.

Développés par la société **Acorn Computers** dans les années 80, la partie création de microprocesseurs de la compagnie est devenue [ARM Ltd](https://www.arm.com/).

ARM propose:

- des CPUs (Cortex)
  - Cortex-A (Smartphone)
  - Cortex-R (Traitement temps réel)
  - Cortex-M( Processeur plus petit : système de contrôle de freinage, le dernier Raspberry Pico)
  - Ethos-N (Machine Learning)
  - Neoverse (serveurs)
- des GPUs (Mali)

ARM se distingue également par son modèle économique : ARM Ltd ne fabrique ni ne vend ses processeurs sous forme de circuits intégrés, elle vend la conception à des partenaires qui les gravent dans le silicium.

Les partenaires de ARM sont principalement les constructeurs de SoC des smartphones:
- Apple
- Qualcomm
- Samsung

Mais également
- Nvidia pour ses cartes graphiques mais également smartphones.
- Ampere Computing, une start up qui a lancé processeurs pour serveur à base d’architecture ARM : les processeurs [Ampere Altra](https://amperecomputing.com/altra/) basés sur les processeurs Neoverse.

### Inconvénients des SoC

Nous avons surtout parlé des avantages du SoC mais il existe évidemment des inconvénients:
- l’extrême intégration de ses composants n’autorise **aucune évolutivité**
- le **recyclage** des composants est problématique
