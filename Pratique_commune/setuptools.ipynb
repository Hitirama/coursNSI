# Setuptools


L'installation de vos programmes passe par la présence d'un fichier nommé **setup.py** à la racine de votre projet.

## Création du fichier setup.py

Ce fichier setup.py permet de déclarer de nombreuses choses :

- Le nom de votre projet
- La description de votre projet
- La version
- L'auteur
- ...

Il permet également de déclarer les dépendances vers des bibliothèques que votre projet utilise.

Enfin et le plus important de tous il installe votre package dans les librairies de Python.

Voici un exemple basique d'un fichier **setup.py**

```python

from setuptools import setup

setup(
    name='le_nom_de_votre_projet',
    version='1',
    description='an implementation of a stack',
    long_description = 'ici une description plus longue',
    author='Mettez votre nom',
    author_email='machin@saint-louis29.net',
    test_suite='le_chemin_vers_le_dossier_des_tests',
    license='Apache Software License',
    packages=['le_nom_de_votre_package'],
    install_requires=['setuptools',],
    zip_safe=False,
)
```

## Installation

Allez dans votre package (à l'endroit où se trouve le fichier setup.py)


Depuis un python virtuel, lancer la commande suivante :

```bash
pip3 install -e .
```

<div class="alert alert-info">L'option -e permet d'être en mode développeur, les modification que vous apportez seront prises en compte immédiatement sans avoir à réinstaller le package.</div>
