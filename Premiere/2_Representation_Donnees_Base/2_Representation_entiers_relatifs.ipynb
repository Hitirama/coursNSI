- On code $+6_{10}=0110_2$:

<div class="alert alert-info"><b>La méthode du complément à deux est l'algorithme le plus utilisé pour la représentation des nombres entiers relatifs.</b></div>

### Généralisation sur k bits

- Sur k bits on peut coder les nombres $\{-2^{k-1},....,2^{k-1}-1\}$
- Le complément à deux d'un entier n est obtenu par le calcul $2^k-\lvert n \rvert$
- Le complément à deux du complément à deux d'un entier n est n : $2^k-(2^k-n) = n$ (l'opposé de l'opposé est bien le nombre.)
- Le nom complet de cette opération est "complément à $2^k$"

### Conversion de décimal à binaire

**Exemple**

Coder -36 en complément à deux sur un octet.

#### Première méthode : utilisation de la formule arithmétique $2^k-\lvert n \rvert$

Arithmétiquement, coder le nombre $n_{10}<0$ en complément à deux revient à coder le nombre $2^k-\lvert n_{10} \rvert$ où k est le nombre de bits utilisés pour l'encodage.


On code donc le nombre $2^8-36=256-36=220_{10}=11011100_2$ (voir le codage binaire d'un entier décimal)

#### Deuxième méthode : Codage en suivant l'algorithme pas à pas

### Conversion de binaire à décimal

**Exemple**

Coder $11011100_2$ en décimal (c'est un nombre entier négatif)

#### Première méthode : utilisation de la formule arithmétique

- Si le bit de poids fort est 1, il s'agit donc d'un nombre négatif
  - On convertit le nombre en décimal : $11011100_2=220_{10}$
  - Le nombre en complément à 2 est donc $2^8-220=256-220=30$
  - Le nombre est donc -30.
- Si le bit de poids fort est 0, il s'agit donc d'un nombre positif
  - On décode 'normalement'

#### Deuxième méthode : Codage en remontant l'algorithme pas à pas

Remonter l'algorithme pour un nombre négatif serait:

- Soustraire 1 : $11011100_2 \to 11011011_2$
- Inverser les bits : $11011011_2 \to 00100100_2$
- Convertir en décimal : $00100100_2 \to 36_{10}$
- Trouver l'opposé  : $36_{10} \to -36_{10}$

Mais une méthode plus facile revient à faire:

- Inverser les bits : $11011100_2 \to 00100011_2$
- Ajouter 1 : $00100011_2 \to 00100100_2$
- Convertir en décimal : $00100100_2 \to 36_{10}$
- Trouver l'opposé : $36_{10} \to -36_{10}$

## Estimation du nombre de bits pour coder un nombre

Nous savons déjà que si nous diposons de k bits on peut coder :

- pour les nombre entiers positifs les nombres de 0 à $2^k-1$. Exemple si k=8 on peut coder de 0 à 256-1=255.
- pour les nombre entiers relatifs les nombres de $-2^{k-1}$ à $2^{k-1}-1$. Exemple si k=8 on peut coder de -128 à +127.

Maintenant posons nous la question inverse : Etant donné un nombre *n*, quel est le nombre de bits *k* nécessaire pour le coder ?


### Première méthode : en tatonnant

**Exemple** : Combien de bits faut-il pour coder $1000000$ au format des entiers naturels ?

- prenons k=10, on obtient $2^{10}-1=1023$
- prenons k=15, on obtient $2^{15}-1=32767$
- prenons k=20, on obtient $2^{20}-1=1048575$
- prenons k=19, on obtient $2^{19}-1=524287$

Il faut donc 20 bits pour coder le nombre 1 million.

**Excercice** combien de bits pour coder 1 milliard au format des entiers naturels?

### Deuxième méthode : avec le logarithme

**La fonction logarithme en base 2 peut être vue comme la fonction qui renvoie la puissance de 2 d'un nombre *n* que l'on écrirait sous la forme $n=2^k$**

**Exemples :**


- $\log_{2}(4)=\log_{2}(2^2) = 2$
- $\log_{2}(256)=\log_{2}(2^8) = 8$
- $\log_{2}(\frac{1}{8})=\log_{2}(2^{-3}) = -3$

Ainsi, pour répondre à la question précédente : 

On cherche k tel que $1000000 = 2^k -1\Leftrightarrow 1000000 +1 = 2^k  \Leftrightarrow k = \log_{2}(1000001) \Leftrightarrow k \approx 19.93$

Il faut donc bien 20 bits pour coder le nombre 1 million.


**Excercice** combien de bits pour coder -36000 au format des entiers relatifs?
